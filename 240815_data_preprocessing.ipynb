# 불필요한 단어 제거

```python
new_train = train.copy()
new_test = test.copy()
```

```python
# '', "", [], .과 같은 특수문자 제거

new_train['문제'] = train['문제'].replace(r'[\"\[\]]','', regex=True)
new_train['선택지'] = train['선택지'].replace(r'[\"\[\]\.]','', regex=True)
new_train.head(15)
```

# 선택지 개수 확인
- ','를 기준으로 확인하려니 한 선택지에 ','가 이미 포함되어 있어서 확인 불가

```python
# 선택지 컬럼에서 ','를 기준으로 몇 개의 단어가 있는지 길이 체크

new_train['선택지 길이'] = new_train['선택지'].str.len()
new_train.head(15)
```

```python
new_train.iloc[0, 1]
```

```python
new_train[new_train['선택지 길이'] >= 5]
```

```python
# 선택지 컬럼에서 중복되는 단어가 있는 데이터 추출
def has_duplicates(lst):
  return len(lst) != len(set(lst))

# 중복되는 단어가 있는 행 추출
duplicate_rows = new_train[new_train['선택지'].apply(has_duplicates)]
duplicate_rows
```

```python
# 중복되는 단어 제거
new_train['선택지'] = new_train['선택지'].apply(lambda x: list(set(x)))

# 선택지 컬럼에서 ','를 기준으로 몇 개의 단어가 있는지 길이 체크
new_train['선택지 길이'] = new_train['선택지'].str.len()
new_train
```
