```python
import pandas as pd
```

```python
df = pd.read_csv('test_data.csv')
df.head()
df['cluster'] = 0
```

```python
df.info()
```

```python
# '문제' 컬럼이 '니까?'로 끝나는 행들의 'cluster' 값을 1로 변경
df.loc[df['문제'].str.endswith('니까?'), 'cluster'] = 1

# '문제' 컬럼이 '시나리오 1 |'을 포함하는 행들의 'cluster' 값을 2로 변경
df.loc[df['문제'].str.contains('시나리오 1 |', regex=False), 'cluster'] = 2

# '문제' 컬럼이 '___'을 포함하는 행들의 'cluster' 값을 3으로 변경 -> 빈공간을 채워야 함
df.loc[df['문제'].str.contains('___', regex=False), 'cluster'] = 3

# '문제' 컬럼이 ':'로 끝나는 'cluster' 값을 4로 변경 -> 답이랑 같이 보고 결정하자!
df.loc[df['문제'].str.endswith(':'), 'cluster'] = 4

# '문제' 컬럼이 특수문자로 끝나지 않는 값을 5로 변경 -> 완성되지 않은 문장을 완성해야 함
df.loc[~df['문제'].str.contains(r'[.,:;!?(){}\[\]\'""]$', regex=True)] = 5
```
