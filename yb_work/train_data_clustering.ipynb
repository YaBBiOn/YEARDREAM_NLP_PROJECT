```python
# '문제' 컬럼이 '니까?'로 끝나는 행들의 'cluster' 값을 1로 변경
df.loc[df['문제'].str.endswith('니까?'), 'cluster'] = 1

# '문제' 컬럼이 '시나리오 1 |'을 포함하는 행들의 'cluster' 값을 2로 변경
df.loc[df['문제'].str.contains('시나리오 1 |', regex=False), 'cluster'] = 2

# '문제' 컬럼이 '___'을 포함하는 행들의 'cluster' 값을 3으로 변경 -> 빈공간을 채워야 함
df.loc[df['문제'].str.contains('___', regex=False), 'cluster'] = 3

# '문제' 컬럼이 ':'로 끝나는 'cluster' 값을 4로 변경 -> 답이랑 같이 보고 결정하자!
df.loc[df['문제'].str.endswith(':'), 'cluster'] = 4

# '문제' 컬럼이 특수문자로 끝나지 않는 값을 5로 변경 -> 완성되지 않은 문장을 완성해야 함
df.loc[~df['문제'].str.contains(r'[.,:;!?(){}\[\]\'""]$', regex=True)] = 5
```

```python
df['len'] = df['문제'].str.len()
len_count = df['len'].value_counts().sort_index()
total = len_count.sum()
prob = len_count / total
print(prob)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10, 6))
prob.plot(kind='bar')
plt.xlabel('문제 길이')
plt.ylabel('확률')
plt.title('문제 길이별 확률')
plt.show
```

```python
# 문장 길이의 사분위수 계산
min_length = df['len'].min()
q1 = df['len'].quantile(0.25)  # 1사분위수
median = df['len'].median()    # 중앙값
q3 = df['len'].quantile(0.75)  # 3사분위수
max_length = df['len'].max()

print(f"문장 길이의 최소값: {min_length}")
print(f"1사분위수 (Q1): {q1}")
print(f"중앙값 (Q2/Median): {median}")
print(f"3사분위수 (Q3): {q3}")
print(f"문장 길이의 최대값: {max_length}")
```
